#### Ques) From API: https://reqres.in/api/users and API: https://reqres.in/api/users?page=2

**Create 2 dataframes respectively with Headers:[ 'id','email','first_name','last_name'] and perform the following operations:**

1. Compare the two DataFrames.

2. Append the two DataFrames and sort the resultant one on basis of First Name (asc order).

3. Form a final dataframe consisting of IDs: 4-10 inclusive.

```python
import pandas as pd
import requests
```

```python
url1 = "https://reqres.in/api/users"
url2 = "https://reqres.in/api/users?page=2"
```

```python
r1 = requests.get(url1)
r1
```

```python
r2 = requests.get(url2)
r2
```

```python
json1 = r1.json()
json1.keys()
json1['data']
```

```python
json2 = r2.json()
json2.keys()
json2['data']
```

```python
df = pd.DataFrame(json1['data'])
df1 = df.drop(['avatar'],axis=1)
df1
```

```python
df = pd.DataFrame(json2['data'])
df2 = df.drop(['avatar'],axis=1)
df2
```

```python
frames = [df1,df2]
df = pd.concat(frames)
df
```

```python
df.sort_values(by=['first_name'],ascending = True)
```

```python
df[3:10]
```
